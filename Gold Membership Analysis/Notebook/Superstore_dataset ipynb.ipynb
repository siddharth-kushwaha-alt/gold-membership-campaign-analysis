```python
channels = {
    "Web": df["NumWebPurchases"].sum(),
    "Store": df["NumStorePurchases"].sum(),
    "Catalog": df["NumCatalogPurchases"].sum()
}
channels
```

```python
import matplotlib.pyplot as plt

plt.bar(channels.keys(), channels.values())

plt.title("Total Purchases by Channel")
plt.xlabel("Channel")
plt.ylabel("Total Purchases")

plt.show()
```

```python
total_web = df["NumWebPurchases"].sum()
total_store = df["NumStorePurchases"].sum()
total_catalog = df["NumCatalogPurchases"].sum()

total = total_web + total_store + total_catalog

print("Web %:", total_web/total*100)
print("Store %:", total_store/total*100)
print("Catalog %:", total_catalog/total*100)
```

```python
df["Deal_Ratio"] = (
    df["NumDealsPurchases"] /
    df["Total_Purchases"]
)
```

```python
df["Discount_Segment"] = pd.cut(
    df["Deal_Ratio"],
    bins=[0,0.3,0.7,1],
    labels=[
        "Low Dependency",
        "Moderate Dependency",
        "High Dependency"
    ]
)
```

```python
corr = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
important_cols = [
    "Response",
    "Income",
    "Age",
    "Recency",
    "Total_Spending",
    "Total_Purchases",
    "NumDealsPurchases"]

corr_subset = df[important_cols].corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr_subset,
    annot=True,
    cmap="coolwarm",
    fmt=".2f")

plt.title("Key Variable Correlations")
plt.show()
```
